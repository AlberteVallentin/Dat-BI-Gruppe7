## JSON

```python
# Import the necessary libraries
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import numpy as np
```

```python
# Set the style for our visualizations
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("deep")
```

```python
# 1. Loading the JSON Data
print("1. Loading the JSON Data")
print("-----------------------")

# Load the senators data from the JSON file
with open('senators.json', 'r') as file:
    senators_data = json.load(file)

# Print basic information about the data
print(f"Meta information: {senators_data['meta']}")
print(f"Total number of senators: {len(senators_data['objects'])}")
print()
```

```python
# 2. Exploring the JSON Structure
print("2. Exploring the JSON Structure")
print("------------------------------")

# Look at the first senator's data to understand the structure
first_senator = senators_data['objects'][0]
print("Keys in the senator object:")
print(sorted(first_senator.keys()))
print("\nKeys in the 'person' sub-object:")
print(sorted(first_senator['person'].keys()))

# Print detailed information about the first senator
print("\nExample senator information:")
print(f"Name: {first_senator['person']['name']}")
print(f"State: {first_senator['state']}")
print(f"Party: {first_senator['party']}")
print(f"Gender: {first_senator['person']['gender_label']}")
print(f"Birthday: {first_senator['person']['birthday']}")
print(f"Phone: {first_senator['phone']}")
print(f"Website: {first_senator['website']}")
print()
```

```python
# 3. Converting to a Pandas DataFrame for easier analysis
print("3. Converting to a Pandas DataFrame")
print("--------------------------------")

# Create a function to flatten the nested JSON structure
def flatten_senator(senator):
    """Extract key information from a senator object into a flat dictionary."""
    return {
        'name': senator['person']['name'],
        'first_name': senator['person']['firstname'],
        'last_name': senator['person']['lastname'],
        'gender': senator['person']['gender_label'],
        'birthday': senator['person']['birthday'],
        'party': senator['party'],
        'state': senator['state'],
        'description': senator['description'],
        'rank': senator['senator_rank'],
        'class': senator['senator_class'],
        'phone': senator['phone'],
        'website': senator['website'],
        'twitter': senator['person'].get('twitterid', None),
        'start_date': senator['startdate'],
        'end_date': senator['enddate']
    }
```

```python
# Apply the function to each senator and create a DataFrame
senators_list = [flatten_senator(senator) for senator in senators_data['objects']]
df_senators = pd.DataFrame(senators_list)
```

```python
# Display the first few rows
print("First 5 rows of the DataFrame:")
print(df_senators.head())
```

```python
# Display the DataFrame info
print("\nDataFrame information:")
print(df_senators.info())
print()
```

```python
# 4. Basic Analysis
print("4. Basic Analysis")
print("---------------")

# 4.1 Party Distribution
print("4.1 Party Distribution")
party_counts = df_senators['party'].value_counts()
print(party_counts)
print()
```

```python
# 4.2 Gender Distribution
print("4.2 Gender Distribution")
gender_counts = df_senators['gender'].value_counts()
print(gender_counts)
print()
```

```python
# 4.3 State Distribution
print("4.3 State Distribution")
state_counts = df_senators['state'].value_counts()
print("Number of senators per state:")
print(state_counts)
print(f"\nTotal states represented: {len(state_counts)}")
print()
```

```python
# 4.4 Calculate Ages
print("4.4 Age Analysis")
# Convert birthday to datetime
df_senators['birthday'] = pd.to_datetime(df_senators['birthday'])

# Calculate age
current_date = datetime.now()
df_senators['age'] = df_senators['birthday'].apply(lambda x: 
    current_date.year - x.year - ((current_date.month, current_date.day) < (x.month, x.day)))

print("Age statistics:")
print(f"Average age: {df_senators['age'].mean():.1f} years")
print(f"Youngest senator: {df_senators['age'].min()} years")
print(f"Oldest senator: {df_senators['age'].max()} years")
print()

# Get the youngest and oldest senators
youngest = df_senators.loc[df_senators['age'].idxmin()]
oldest = df_senators.loc[df_senators['age'].idxmax()]

print(f"Youngest senator: {youngest['name']}, Age: {youngest['age']}, Party: {youngest['party']}")
print(f"Oldest senator: {oldest['name']}, Age: {oldest['age']}, Party: {oldest['party']}")
print()
```

```python
# 4.5 Senate Classes
print("4.5 Senate Classes")
class_counts = df_senators['class'].value_counts()
print(class_counts)
print()
```

```python
# 4.6 Senate Ranks
print("4.6 Senate Ranks")
rank_counts = df_senators['rank'].value_counts()
print(rank_counts)
print()
```

```python
# 5.1 Party and Gender Relationship
print("5.1 Party and Gender Relationship")
party_gender = pd.crosstab(df_senators['party'], df_senators['gender'])
print(party_gender)
print()
```

```python
# Calculate percentages
print("Gender percentage within each party:")
party_gender_pct = party_gender.div(party_gender.sum(axis=1), axis=0) * 100
print(party_gender_pct)
print()
```

```python
# 5.3 Age Distribution by Party
print("5.3 Age Statistics by Party")
age_by_party = df_senators.groupby('party')['age'].agg(['mean', 'min', 'max']).round(1)
print(age_by_party)
print()
```

```python
# 5.4 Term Analysis
print("5.4 Term Analysis")
# Convert dates to datetime
df_senators['start_date'] = pd.to_datetime(df_senators['start_date'])
df_senators['end_date'] = pd.to_datetime(df_senators['end_date'])

# Calculate current term length in years
df_senators['term_length'] = (pd.to_datetime(df_senators['end_date']) - 
                             pd.to_datetime(df_senators['start_date'])).dt.days / 365.25

print("Term length statistics (years):")
print(f"Average term length: {df_senators['term_length'].mean():.2f} years")
print(f"Shortest term: {df_senators['term_length'].min():.2f} years")
print(f"Longest term: {df_senators['term_length'].max():.2f} years")
print()
```

```python
# 6.1 Party Distribution Pie Chart
plt.figure(figsize=(10, 6))
plt.pie(party_counts, labels=party_counts.index, autopct='%1.1f%%', startangle=90,
        colors=['red', 'blue', 'green'])
plt.title('US Senate Composition by Party', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show() 
```

```python
# 6.2 Gender Distribution Bar Chart
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Gender Distribution in the US Senate', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Senators', fontsize=12)

# Add value labels on top of the bars
for i, v in enumerate(gender_counts.values):
    ax.text(i, v + 0.5, str(v), ha='center', fontsize=12)

plt.tight_layout()
plt.show()  
```

```python
# 6.3 Age Distribution Histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=df_senators, x='age', bins=10, kde=True)
plt.title('Age Distribution of US Senators', fontsize=16)
plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Number of Senators', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axvline(df_senators['age'].mean(), color='red', linestyle='dashed', linewidth=1, 
           label=f'Mean Age: {df_senators["age"].mean():.1f}')
plt.legend()
plt.tight_layout()
plt.show() 
```

```python
# 6.4 Age by Party Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_senators, x='party', y='age')
plt.title('Age Distribution by Party', fontsize=16)
plt.xlabel('Party', fontsize=12)
plt.ylabel('Age (years)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 6.5 Party and Gender Grouped Bar Chart
plt.figure(figsize=(12, 6))
party_gender.plot(kind='bar')
plt.title('Gender Distribution by Party', fontsize=16)
plt.xlabel('Party', fontsize=12)
plt.ylabel('Number of Senators', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
# 7.1 Find senators from a specific state
print("7.1 Senators from California:")
california_senators = df_senators[df_senators['state'] == 'CA']
print(california_senators[['name', 'party', 'rank']])
print() 
```

```python
# 7.2 Find female Republican senators
print("7.2 Female Republican Senators:")
female_republicans = df_senators[(df_senators['party'] == 'Republican') & 
                                (df_senators['gender'] == 'Female')]
print(female_republicans[['name', 'state']])
print()
```

```python
# 7.3 Find the oldest senator from each party
print("7.3 Oldest Senator from Each Party:")
oldest_by_party = df_senators.loc[df_senators.groupby('party')['age'].idxmax()]
print(oldest_by_party[['name', 'party', 'age']])
print()
```

```python
# 7.4 Find senators under 50 years old
print("7.4 Senators Under 50 Years Old:")
young_senators = df_senators[df_senators['age'] < 50]
print(young_senators[['name', 'party', 'age']].sort_values('age'))
print()
```

```python
# 8.1 Convert DataFrame back to JSON
print("8.1 Converting DataFrame back to JSON")
df_json = df_senators.to_json(orient='records', indent=2)
print("First 200 characters of the JSON:")
print(df_json[:200] + "...")
print()
```

```python
# 8.2 Create a modified JSON object
print("8.2 Creating a Modified JSON Structure")
# Create a simplified JSON with just name, state, and party
simplified_senators = []
for senator in senators_data['objects']:
    simplified_senators.append({
        'name': senator['person']['name'],
        'state': senator['state'],
        'party': senator['party']
    })

# Convert to JSON string
simplified_json = json.dumps(simplified_senators, indent=2)
print("First 200 characters of the simplified JSON:")
print(simplified_json[:200] + "...")
print()
```

```python
# 8.3 Save the simplified JSON to a file
print("8.3 Saving JSON to a File")
with open('simplified_senators.json', 'w') as file:
    json.dump(simplified_senators, file, indent=2)
print("Saved simplified_senators.json")
print()
```
