#### Part 1: Load the data

```python
print("Loading wine quality datasets...")

# Now we need to load the dataset into a pandas DataFrame
dataset_redwine = 'Dataset/winequality-red.xlsx'

# We make a function to load the dataset
def load_xlsx_to_dataframe(file_path):
    df = pd.read_excel(file_path, header=1)
    return df

# And transform it into a DataFrame
df_redwine = load_xlsx_to_dataframe(dataset_redwine)

# Print the dataframe
df_redwine
```

### Aggregation of the two data sources

```python
# File paths
dataset_redwine = 'Dataset/winequality-red.xlsx'
dataset_whitewine = 'Dataset/winequality-white.xlsx'

# Read both datasets using read_excel
red_df = pd.read_excel(dataset_redwine, header=1)
white_df = pd.read_excel(dataset_whitewine, header=1)

# Add a new column to distinguish types
red_df["wine_type"] = "red"
white_df["wine_type"] = "white"

# Combine both datasets
wine_df = pd.concat([red_df, white_df], ignore_index=True)

# Optional: check the combined dataset
wine_df
```
